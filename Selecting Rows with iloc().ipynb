# Selecting Rows with iloc()

##  ----  Quick Summary ----
## What are major import topics features?
### 1. iloc[0], first
### 2. iloc[1], second, iloc[-1] last
### 3. iloc[[0,1,2]] or iloc[0:3]
### 4. iloc[:5] first 5 row
### 5. iloc[-5:] last 5 row

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head(5)
```

```python
summer = pd.read_csv("summer.csv",index_col="Athlete")
```

```python
summer.head(5)
```

## iloc[0] means first row

```python
summer.iloc[0]
```

## iloc[1] means second row

```python
summer.iloc[1]
```

## iloc[-1] means last row

```python
summer.iloc[-1]
```

## iloc[ [1, 2, 3] ] means first 3 row

```python
summer.iloc[[0,1,2]]
```

```python
summer.iloc[0:3]
```

# first 5 row

```python
summer.iloc[:5]
```

# last 5 row

```python
summer.iloc[-5:]
```

# 2nd. 45th. 5467th.

```python
summer.iloc[[2,45,5467]]
```
