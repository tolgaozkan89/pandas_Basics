# Slicing Rows & Columns with loc()

##  ----  Quick Summary ----
## What are major import topics features?
### 1. iloc[0], first
### 2. iloc[1], second, iloc[-1] last
### 3. iloc[[0,1,2]] or iloc[0:3]
### 4. iloc[:5] first 5 row
### 5. iloc[-5:] last 5 row

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head(5)
```

```python
summer = pd.read_csv("summer.csv",index_col="Athlete")
```

```python
summer.head(5)
```

## loc["PHELPS, Michael"] directly finds the swimmer

```python
summer.loc["PHELPS, Michael"]
```

## loc["PHELPS, Michael","Medal"] directly finds the medals of Phelps

```python
summer.loc["PHELPS, Michael","Medal"]
```

## loc["PHELPS, Michael","Medal","Event"] ERROR

```python
summer.loc["PHELPS, Michael","Medal","Event"]
```

## loc["PHELPS, Michael", ["Medal","Event"] ] CORRECT

```python
summer.loc["PHELPS, Michael",["Medal","Event"]]
```

## loc[ [ "PHELPS, Michael", "LEWIS, Carl" ] ["Medal","Event"] ] 2 swimmer + 2 feature

```python
summer.loc[["PHELPS, Michael","LEWIS, Carl"],["Medal","Event"]]
```

## loc[ : , ["Medal","Event"] ]  All rows of Medal and Event

```python
summer.loc[:,["Medal","Event"]]
```
